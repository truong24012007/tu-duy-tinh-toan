W3A1:Viết một chương trình nhập vào một số nguyên n và in ra các chữ số đảo ngược nhau

```python
n = int(input('nhập vào số nguyên n'))
dao = 0
temp = abs(n)
while temp > 0:
    dao = dao * 10 + temp % 10
    temp //= 10
if n < 0:
    dao = -dao
print('số đảo ngược', dao)        
```

W3A3: Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0 


```python
n = int(input("Nhập số nguyên n: "))
if n > 0 and (n & (n - 1)) == 0:
    print(n, "là lũy thừa của 2")
else:
    print(n, "không phải là lũy thừa của 2")
```

W3A11: Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp
lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay
"Tam giác thường".

```python
a, b, c = map(float, input().split())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print('Tam giác đều')
    elif a == b or a == c or b == c:
        print('Tam giác cân')     
    elif a ** 2 == b ** 2 + c ** 2 or b ** 2 == a ** 2 + c ** 2 or c ** 2 == b ** 2 + a ** 2:
        print('Tam giác vuông')
    else:
        print('Tam giác thường')     
else:
    print('Không phải tam giác')              
```

W3A13: Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh

```python
sodien = float(input())
if 0 <= sodien <= 50:
   tien = sodien * 1500
   print(tien)
elif 51 <= sodien <= 100:
   tien = 50 * 1500 + (sodien - 50) * 2000
   print(tien)
else:
   tien = 50 * 1500 + 50 * 2000 + (sodien - 100) * 3000   
   print(tien)
```

W3A17: Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng
thứ tự a, b, c, d hay không?
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội và so sánh vs c, c nhân công bội và so sánh vs
d. 

```python
a, b, c, d = map(int, input().split())
q = b / a
if b % a == 0 and c == b * q and d == c * q:
    print('Yes')
else:
    print('No')    
```
