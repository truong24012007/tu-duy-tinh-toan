W5A16: Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_amstronhg(x):
    s = str(x)
    k = len(s)
    tong = 0
    for i in s:
        tong += int(i) ** k
    return tong == x
if __name__ == '__main__':
    x = int(input()) 
    if is_amstronhg:
        print('Yes')
    else:
        print('No')       
```

W5A17: Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_cp(x):
    if x < 0:
        return False
    i = 0
    while i * i <= x:
        if i * i == x:
            return True
        i += 1
    return False    
if __name__ == '__main__':
    x = int(input())
    if is_cp(x):
        print('Yes')
    else:
        print('No')    
```

W5A18: VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def is_dx(x):
    if x < 0:
        return False
    strart = x
    rev = 0
    while x != 0:
        rev = rev * 10 + x % 10
        x //= 10
    return strart == rev
if __name__ == '__main__':
    x = int(input())
    if is_dx(x):
        print('True')
    else:
        print('False')        
     
```

W5A19: VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_uoc(x):
    tong = 0
    i = 1
    while i < x:
        if x % i == 0:
            tong += i
        i += 1
    return tong == x
if __name__ == '__main__':
    x = int(input())
    if is_uoc(x):
        print('Là số kì diệu')
    else:
        print('Không là số kì diệu')        
```

W5A20: Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except:
        try:
            float(s)
            return 'float'
        except:
            return 'neither'
```

W5A24: Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    return (x2 - x1) * (y3 - y1) == (y2 - y1) * (x3 - x1)
```

W5A25: Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)


```python
def cross(x1, y1, x2, y2, x3, y3):
    return (x2 - x1)*(y3 - y1) - (y2 - y1)*(x3 - x1)
def point_in_triangle(A, B, C, P):
    x1, y1 = A
    x2, y2 = B
    x3, y3 = C
    x, y = P
    d1 = cross(x, y, x1, y1, x2, y2)
    d2 = cross(x, y, x2, y2, x3, y3)
    d3 = cross(x, y, x3, y3, x1, y1)

    has_neg = (d1 < 0) or (d2 < 0) or (d3 < 0)
    has_pos = (d1 > 0) or (d2 > 0) or (d3 > 0)

    return not (has_neg and has_pos)  
def point_in_quadrilateral(quad, p):
    A, B, C, D = quad
    return point_in_triangle(A, B, C, p) or point_in_triangle(A, C, D, p)
```

W5A27: Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count_non_zero_digits(nums):
    count = 0
    for n in nums:
        n = abs(n) 
        if n == 0:
            continue
        while n > 0:
            digit = n % 10
            if digit != 0:
                count += 1
            n //= 10
    return count
```

W5A28: Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
def angle_ABC(x1, y1, x2, y2, x3, y3):
    # vector BA và BC
    BAx, BAy = x1 - x2, y1 - y2
    BCx, BCy = x3 - x2, y3 - y2
    dot = BAx * BCx + BAy * BCy
    len_BA = math.hypot(BAx, BAy)
    len_BC = math.hypot(BCx, BCy)
    if len_BA == 0 or len_BC == 0:
        return None
    cos_theta = dot / (len_BA * len_BC)
    cos_theta = max(-1, min(1, cos_theta))
    angle = math.degrees(math.acos(cos_theta))
    return angle
```

W5A30: Viết hàm tính hệ số nhị thức C(n, k).

```python
import math
def C(n, k):
    if k < 0 or k > n:
        return 0
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))
```

W5A33: Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm và giảm giá

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # Giá theo loại xe
    if car_type == "standard":
        base = 10000
        price_10km = 12000
        price_after = 10000
    elif car_type == "premium":
        base = 15000
        price_10km = 15000
        price_after = 12000
    else:
        return None
    if km <= 10:
        fare_km = km * price_10km
    else:
        fare_km = 10 * price_10km + (km - 10) * price_after
    wait_fee = wait_minutes * 2000
    total = base + fare_km + wait_fee
    if start_hour >= 22 or start_hour < 6:
        total *= 1.2
    if km > 20:
        total *= 0.9
    return int(total)
```

W5A34: Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def has_unique_digits(x):
    s = str(x)
    return len(s) == len(set(s))
def sum_1_to_n(n):
    if n <= 0:
        return 0
    total = 0
    for i in range(1, n + 1):
        if has_unique_digits(i):
            total += i
    return total
```
