```python
while True:
    n = int(input())
    if n > 0:
        break
la_snt = True
if n < 2:
     la_snt = False
else: 
    for i in range(2, n - 1):
        if n % i == 0:
            la_snt = False
            break  
if la_snt:
    print("La so nguyen to")
else:
    print("Khong phai so nguyen to")
```

W4A6: Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n - 1):
        if n % i == 0:
            return False
    return True
while True:
    try:
        a, b = map(int, input().split())
        
        tong = 0
        for i in range(a, b + 1):
            if is_prime(i):
                tong += i
        
        print(tong)
    except:
        break
```

W4A7: Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n

```python
n = int(input())
max_nt = 0
i = 2
while n > 1:
    if n % i == 0:
        max_nt = i
        n //= i
    else:
        i += 1
print(max_nt)                     
```

W4A8: Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n = int(input())
steps = 0
while str(n) != (str(n)[::-1]):
     rev = int(str(n)[::-1])
     n = n + rev
     steps += 1
print(steps, n)
```

W4A9: Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number).

```python
n = int(input())
found = False
i = 1
while i * i <= n:
    x = i * i
    s = str(x)
    if len(s) == len(set(s)):
        print(x, end = " ")
        found = True
    i += 1
if not found:
    print('(no number)')        
```

W4A10: Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất."

```python
n = int(input())
max_len = 0
res_x = 1
for x in range(1, n + 1):
    temp = x
    lenght = 1
    while temp != 1:
        if temp % 2 == 0:
          temp = temp // 2
        else:
          temp = temp * 3 + 1
        lenght += 1
    if lenght > max_len:
       max_len = lenght
       res_x = x
print(res_x, max_len)                  
```

W4A11: Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6.

W4A13: Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false" 

```python
a = int(input())
b = int(input())
for i in range(1, a):
    s = 0
    if a % i == 0:
        s += i
tong_a = s     
for j in range(1, b):
    if b % j == 0:
        k = 0
        k += j
tong_b = k     
if tong_a == tong_b:
    print('true')
else:
    print('false')     
```

W4A14: Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
tìm ước chung lớn nhất của hai số m và n. 

```python
m = int(input())
n = int(input())
ucln = 1
for i in range(min(m, n), 0, -1):
    if m % i == 0 and n % i == 0:
        ucln = i
        break
print(ucln)    
```
