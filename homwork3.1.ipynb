Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input())
for i in range(1, h + 1):
    for j in range(h - i):
        print(" ", end = "")
    for j in range(2 * i - 1):
        if i == 1 or i == h or j == 0 or j == 2 * i - 2:
            print("*", end = "")
        else:
            print(" ", end = "")  
    print()              
```

Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 

```python
import math
a = int(input())
if int(math.sqrt(a)) ** 2 == a:
    print('số chính phương')
else:
    print('không phải số chính phương')    
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 


```python
a = int(input())
s = 0
for i in range(1, a):
    if a % i == 0:
        s += i
if s == a:
    print('số hoàn hảo') 
else:
    print('không phải số hoàn hảo')           
```

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)
lớn nhất nhưng nhỏ hơn n 

```python
n = int(input())
S = 0
k = 0
while S + (k + 1) < n:
    k += 1
    S += k
print('k = ', k)
print('S(k) = ', S)    
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
A = int(input())
n = 0
S = 0
while S <= A:
    n += 1
    S += 1 / n
print(n)    
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 

```python
max_val = None
min_val = None
while True:
    x = int(input())
    if x == -1:
        break
    if max_val is None or x > max_val:
        max_val = x
    if min_val is None or x < min_val:
        min_val = x
if max_val is None:
    print('không có số nào được nhập')      
else:
    print("Số lớn nhất:", max_val)
    print("Số nhỏ nhất:", min_val)          
```

Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều
cách nhất có thể, ko dùng mảng. 

```python
n = int(input())
dem = 0
if n == 0:
    dem = 1
else:
    dem = 0    
    for i in range(100):
        if n == 0:
            break
        dem += 1
        n //= 10
print('số chữ số', dem)        
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ 

```python
n = int(input())
dem_chan = 0
dem_le = 0
if n == 0:
    dem_chan = 1
    dem_le = 0
else:
   while n > 0:
        digit = n % 10
        if digit % 2 == 0:
            dem_chan +=1
        else:
            dem_le +=1
        n //= 10    
print("Số chữ số chẵn:", dem_chan)
print("Số chữ số lẻ:", dem_le)                  
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay
không 

```python
n = int(input())
if n <= 0:
    print('không phải')
else:
    while n % 3 == 0:
        n //= 3
    if n == 1:
       print('có dạng 3 ^ k')
    else:
       print('không phải')    


```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A = int(input())
a1, a2 = 1, 1
if A <= 1:
    print('không có dãy')
else:
    while a2 <= A:
        a1, a2 = a2, a1 + a2   
    print("Số Fibonacci lớn nhất không vượt quá A là:", a1)
    
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố. 

```python
n = int(input())
dem = 0
if n == 2:
    dem = 0
else:
    while n > 0:
        digit = n % 10
        if digit in (2, 3, 5, 7):
           dem += 1
        n //= 10
print("Số chữ số nguyên tố là:", dem)          
```
