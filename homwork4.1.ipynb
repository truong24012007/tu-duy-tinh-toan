W4A18: Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a, b = map(int, input().split())
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        print(i, end = ' ')
```

W4A20: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input())
while n > 1:
    if n % 2 != 0:
        print('False')
        break
    n = n // 2
else:
    print('True')
```

W4A23:

Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 

```python
n = int(input())
S = 0
k = 0
while S + (k + 1) < n:
    k += 1
    S += k
print(k)
```

1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

W4A25:
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
x = int(input())
max_val = x
min_val = x
while x != -1:
    if x > max_val:
        max_val = x
    if x < min_val:
        min_val = x
    x = int(input())

print(max_val, min_val)
```

W4A26:
[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 


```python
A = int(input())
a1, b1 = 1, 1
while b1 <= A:
    a1, b1 = b1, a1 + b1
print(a1)    
```

W4A27:
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
Dùng cách: for

```python
s = input()
dem = 0
trong_tu = False
for ch in s:
    if ch != ' ' and trong_tu == False:
        dem += 1
        trong_tu = True
    elif ch == ' ':
        trong_tu = False
print(dem)        
    
```

W4A28
Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi
Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
s = input()
tu = ''
bat_dau = False
for i in range(len(s)):
    if s[i] != ' ':
        bat_dau = True
        tu += s[i]
    elif bat_dau:
        break
print(tu)    
```

W4A29: Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
thường và thường thành hoa. 

```python
s = input()
kq = ''
for ch in s:
    if 'a' <= ch <= 'z':
        kq += chr(ord(ch) - 32)
    elif 'A' <= ch <= 'Z':
        kq += chr(ord(ch) + 32)    
print(kq)        
```

W4A30: Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính
tổng 3 số nguyên đó 

```python
s = input()
tong = 0
so = ""
for ch in s:
    if ch != ',':
        so += ch
    else:
        tong += int(so)
        so = ""
tong += int(so)
print(tong)
```

W4A31: Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số 

```python
s = input()
hoa = 0
thuong = 0
so = 0
for ch in s:
    if 'A' <= ch <= 'Z':
        hoa += 1
    elif 'a' <= ch <= 'z':
        thuong += 1
    elif '0' <= ch <= '9':
        so += 1
print(hoa, thuong, so)
```
